```python
import pandas as pd
df = pd.read_excel(r"C:\Users\ryzen\OneDrive\Alex the analyst data analyst playlist practice\Python\Data cleaning\Customer Call List.xlsx")
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column") # drop non useful columns before cleaning data, could be already deleted
```

```python
df
```

```python
df["Last_Name"].str.lstrip("...") # cannot use [] for multiple
```

```python
df["Last_Name"] = df["Last_Name"].str.lstrip("...")
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("_")
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("/")
```

```python
df
```

```python
# can do strip("123_.") to remove any numbers, _, and .
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True) # replace everything that is not a-zA-Z0-9 with nothing
# remove all special characters to standardize and then can put in the -'s
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)) # for x in df["Phone_Number"] str(x) something like that

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10]) # phone number col needs to be phone number
```

```python
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("nan--", "")
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace("Na--", "")
```

```python
df
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',', n=2, expand=True) # have to have these arguments
```

```python
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")
```

```python
df
```

```python
df = df.fillna('') # if no value, then fill with ""
df = df.replace("N/a", "") # N/a is a value
```

```python
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
```

```python
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y': # loc[Row, Col]
        df.drop(x, inplace=True)
    elif df.loc[x, "Phone_Number"] == '': # loc[Row, Col]
        df.drop(x, inplace=True)

df

# df = df.dropna(subset="Phone_Number", inplace=True) # drops rows with empty phone numbers
```
